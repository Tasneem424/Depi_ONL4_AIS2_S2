```python
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
x=np.array([[1],[2],[3],[4]])
y=np.array([2,4,6,8])
```

```python
lin_reg= LinearRegression()
lin_reg.fit(x,y)
y_pred_lin= lin_reg.predict(x)
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
lin_ridge = Ridge(alpha=10)
lin_ridge.fit(x,y)
y_pred_ridge= lin_ridge.predict(x)
print(lin_ridge.coef_)
print(lin_ridge.intercept_)
```

```python
lin_lasso = Lasso(alpha=10)
lin_lasso.fit(x,y)
y_pred_lasso= lin_lasso.predict(x)
print(lin_lasso.coef_)
print(lin_lasso.intercept_)
```

```python
print("Linear Regression")
print("mse :",mean_squared_error(y,y_pred_lin))
print("r2_score :",r2_score(y,y_pred_lin))
print("========================")
print("Ridge")
print("mse :",mean_squared_error(y,y_pred_ridge))
print("r2_score :",r2_score(y,y_pred_ridge))
print("========================")
print("Lasso")
print("mse :",mean_squared_error(y,y_pred_lasso))
print("r2_score :",r2_score(y,y_pred_lasso))
print("========================")
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.scatter(x,y,color='blue',label="actual data")
plt.plot(x,y_pred_lin,color='red',label="Linear Regression")
plt.title("Linear Regression")
plt.legend()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.scatter(x,y,color='blue',label="actual data")
plt.plot(x,y_pred_ridge,color='red',label="Ridge")
plt.title("Linear Ridge")
plt.legend()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.scatter(x,y,color='blue',label="actual data")
plt.plot(x,y_pred_lasso,color='red',label="Lasso")
plt.title("Linear Regression")
plt.legend()
```
